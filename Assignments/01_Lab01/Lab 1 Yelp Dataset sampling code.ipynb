```python
import pandas as pd
```

```python
reviews_dataset = pd.read_csv('Yelp_Reviews.csv')
```

```python
reviews_dataset
```

```python
len(reviews_dataset)
```

```python
unique_users = reviews_dataset.user_id.unique().tolist()
```

```python
len(unique_users)
```

```python
unique_users = unique_users[int(0.25*686556):int(0.5*686556)]
```

```python
unique_users_df = pd.DataFrame(unique_users,columns=['user_id'])
```

```python
unique_users_df
```

```python
reviews_subset = pd.merge(reviews_dataset, unique_users_df, on='user_id', how='inner')
```

```python
reviews_subset
```

```python
cool_votes = []
useful_votes = []
funny_votes = []
for item in reviews_subset.votes:
    votes = [int(s[:-1]) for s in item.split() if s[:-1].isdigit()]
    cool_votes.append(votes[0])
    useful_votes.append(votes[1])
    funny_votes.append(votes[2])
```

```python
reviews_subset['cool_votes']=cool_votes
reviews_subset['useful_votes'] = useful_votes
reviews_subset['funny_votes'] = funny_votes
```

```python
reviews_subset
```

```python
del reviews_subset['votes']
```

```python
reviews_subset.to_csv('Yelp_Reviews.csv', index=False)
```
