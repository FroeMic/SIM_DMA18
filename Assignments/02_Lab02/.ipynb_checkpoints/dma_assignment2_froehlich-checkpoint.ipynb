### Question 2 ###
What is the best choice of k according to the silhouette metric for clustering q4-q6 (# of cool, funny, useful votes). Only consider 2 <= k <= 8.  

NOTE: For features with high variance, empty clusters can occur. There are several ways of dealing with empty clusters. A common approach is to drop empty clusters, the prefered approach for this Lab is to treat the empty cluster as a “singleton” leaving it empty with a single point placeholder. This can be accomplished in MATLAB, for example, with the command:  [IDX,C]=kmeans(d,4, 'Distance', 'sqeuclidean','EmptyAction','singleton')

**Answer**:  

```python
# q4 :: Average number of cool votes.
# q5 :: Average number of funny votes.
# q6 :: Average number of useful votes.


from sklearn.metrics import silhouette_score
from sklearn.model_selection import train_test_split

# randomly sample about 10,000 instances
X = train_test_split(np.array(df[['q4','q5','q6']]), test_size=0.06)[1]

results = []
for i in range(2,9):
    kmeans = KMeans(n_clusters=i, random_state=0).fit(X)
    score = silhouette_score(X, kmeans.labels_)

    print('[' + str(i) + '] :: ', score)
    results.append((i, score))
```

```python
def select_max(results):
    (best_k, best_score) = 0,0
    for (k, s) in results:
        if s > best_score:
            best_k = k
            best_score = s
    return (best_k, best_score)

best_k, best_score = select_max(results)
print(' Question 2 :: The best choice of k is', best_k, 'with a silhouette score of', best_score)

# Note:
# Could you follow up in one of the lectures, which kind of business questions are likely to be answered by clustering?
# I am not sure, what I information learned from creating these clusterings.
# Is is just an initial step to better understand one's data, or what do I get out of having a clustering?
#
# Also, when the best clustering (according to the silhoutte score) is achieved with k=2 
# and we only have 3 input features, is k-means really the way to go?
# With such a simple input space, couldn't rule based clustering be more accurate and not
# limited in the same way as k-means (e.g. only spherial clusters)
```

### Question 5 ###
Inspect the [best] clustering generated from question 4

#### 7.a ####
**Question**: Using the silhouette metric, what was the best k?  

**Answer**:  

```python
# print(len(X_norm)) #65546

X_clust = train_test_split(np.array(X_norm), test_size=0.16)[1]

results = []
for i in range(2,9):
    kmeans = KMeans(n_clusters=i, random_state=0).fit(X_clust)
    score = silhouette_score(X_clust, kmeans.labels_)

    print('[' + str(i) + '] :: ', score)
    results.append((i, score))
```

```python
best_k, best_score = select_max(results)
print(' Question 7a :: The best choice of k is', best_k, 'with a silhouette score of', best_score)
```

#### 7.b ####
**Question**: Using the the sum of within cluster variance metric with the elbow method what was the best k?  
**Answer**:  

```python
import scipy

results = []
for k in range(2,9):
    kmeans = KMeans(n_clusters=k, random_state=0).fit(X_clust)
    err = 0
    for i, c in enumerate(kmeans.labels_):
        p = X_clust[i]
        c_i = kmeans.cluster_centers_[c]
        dist = scipy.spatial.distance.euclidean(p, c_i)
        err += np.power(dist, 2)

    results.append(err)
    print('k=' + str(k), ' :: ',  err)
    
  

plt.plot(list(range(2,9)), results)
plt.show()

# Where is the elbow? 
```
