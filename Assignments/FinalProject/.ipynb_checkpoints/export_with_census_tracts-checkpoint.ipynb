```python
import os

def make_sure_directory_exists(path):
    if not os.path.exists(path):
        os.makedirs(path)
        
def export(path, filename, df):
    make_sure_directory_exists(path)
    filepath = os.path.join(path, filename)
    df.to_csv(filepath ,index=False,, sep=',')
    
```
